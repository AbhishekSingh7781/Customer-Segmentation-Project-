```python
# Bivariate Analysis
plt.figure(figsize=(6,4))
sns.scatterplot(x='Annual Income(USD)', y='Spending Score (1-100)', data=df)
plt.title('Income vs Spending Score')
plt.show()
```

```python
sns.boxplot(x='Gender', y='AvgPurchaseValue (USD)', data=df)
plt.title('Gender vs Avg Purchase Value')
plt.show()
```

```python
sns.scatterplot(x='Age', y='Spending Score (1-100)', data=df)
plt.title('Age vs Spending Score')
plt.show()
```

```python
sns.scatterplot(x='Purchase Frequency(per month)', y='LoyaltyScore (1-10)', data=df)
plt.title('Purchase Frequency vs Loyalty Score')
plt.show()
```

```python
sns.barplot(x='Preferred Category', y='AvgPurchaseValue (USD)', data=df, estimator='mean',ci=None)
plt.title('Category vs Avg Purchase Value')
plt.xticks(rotation=45)
plt.show()
```

```python
sns.barplot(x='Tenure (Years)', y='LoyaltyScore (1-10)', data=df, estimator='mean', ci=None)
plt.title("Tenure vs Loyalty Score")
plt.show()
```

```python
#Correlation Heatmap
plt.figure(figsize=(10,6))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap="coolwarm", fmt=".2f")
plt.title('Correlation Heatmap')
plt.show()
```
